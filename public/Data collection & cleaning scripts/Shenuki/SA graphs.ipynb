```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data from the Excel file
input_file = 'Reddit_AU_SA_cleaned.xlsx'
df = pd.read_excel(input_file)

# Remove rows where the "Content" column is empty
df = df.dropna(subset=['Content'])

# Plot the distribution of sentiment categories
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Sentiment Category', palette='coolwarm')
plt.title('Distribution of Sentiment Categories for Reddit Posts on Renewable Energy in Australia')
plt.xlabel('Sentiment Category')
plt.ylabel('Count')
plt.show()

# Plot the sentiment polarity distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Sentiment'], bins=20, kde=True, color='skyblue')
plt.title('Sentiment Polarity Distribution for Reddit Posts')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```
